```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time

nome_filme = 'vingadores'
chrome_options = Options()
chrome_options.page_load_strategy = 'eager'
driver = webdriver.Chrome(options=chrome_options)

linkurl = f"https://www.adorocinema.com/pesquisar/?q={nome_filme}"
print(f"Acessando: {linkurl}")

try:
    driver.get(linkurl)
    time.sleep(3)

    # 1. Seleciona e clica no primeiro filme
    elemento_js = driver.find_element(By.CSS_SELECTOR, "section.movies-results .meta-title-link")
    driver.execute_script("arguments[0].click();", elemento_js)
    print("Sucesso: Clique efetuado via JS!")
    
    print("Aguardando 5 segundos para a página do filme carregar...")
    time.sleep(5)

    # 2. Extração de Géneros
    print("Carregando Gêneros...")
    container = driver.find_element(By.CLASS_NAME, "meta-body-info")
    elementos_generos = container.find_elements(By.CSS_SELECTOR, "a.dark-grey-link")
    generos = [g.text.strip() for g in elementos_generos]

    # 3. Extração de Diretores
    print("Carregando Diretores...")
    container2 = driver.find_element(By.CLASS_NAME, "meta-body-oneline")
    elementos_diretores = container2.find_elements(By.CSS_SELECTOR, "a.dark-grey-link")
    diretores = [d.text.strip() for d in elementos_diretores]

    # 4. Extração do Link da Imagem (Poster)
    # Dica: No AdoroCinema, o poster costuma estar dentro desta classe
    print("Carregando link da imagem...")
    imagem_elemento = driver.find_element(By.CSS_SELECTOR, "figure.thumbnail img")
    link_da_imagem = imagem_elemento.get_attribute("src")

    # Exibição dos Resultados
    print("-" * 30)
    print(f"Gêneros: {generos}")
    print(f"Diretores: {diretores}")
    print(f"Link da imagem: {link_da_imagem}")
    print("-" * 30)

except Exception as e:
    print(f"Erro durante a execução: {e}")

finally:
    driver.quit() # O quit() já fecha todas as janelas, não precisas do close() antes
```
