```python
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
chrome_options = Options()
# chrome_options.add_argument("--headless") 
chrome_options.page_load_strategy = 'eager'
driver = webdriver.Chrome(options=chrome_options)
driver.get("https://www.adorocinema.com/filmes/filme-232669/")
```

```python
container = driver.find_element(By.CLASS_NAME, "meta-body-info")
elementos_generos = container.find_elements(By.CSS_SELECTOR, "a.dark-grey-link")
generos = [g.text.strip() for g in elementos_generos]
print(f"Gêneros encontrados: {generos}")

# driver.close() # Fecha a janela atual (aba) que o Selenium está controlando
# driver.quit()  # Encerra o processo do driver e do navegador por completo
```

```python
container2 = driver.find_element(By.CLASS_NAME, "meta-body-oneline")
elementos_diretores = container2.find_elements(By.CSS_SELECTOR, "a.dark-grey-link")
diretores = [g.text.strip() for g in elementos_diretores]
print(f"Diretores encontrados: {diretores}")

# driver.close() # Fecha a janela atual (aba) que o Selenium está controlando
# driver.quit()  # Encerra o processo do driver e do navegador por completo
```

```python
imagem_elemento = driver.find_element(By.CLASS_NAME, "thumbnail-img")
link_da_imagem = driver.find_element(By.CSS_SELECTOR, "a.thumbnail-link img").get_attribute("src")
print(f"Link da imagem: {link_da_imagem}") 
```
